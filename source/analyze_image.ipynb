### 載入各種工具包

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
```

### 照片位置   目前為相對位置

```python
file_name = "picture/test2.jpg"
```

### 先顯示照片看看   因為彩色圖片為ＲＧＢ共三層，分層顯示

```python
def mat_version(file_name):
    img = mpimg.imread(file_name)
    lum_img = img[:, :, 0]

    # This is array slicing.  You can read more in the `Numpy tutorial
    # <https://docs.scipy.org/doc/numpy-dev/user/quickstart.html>`_.
    
    fig, axs = plt.subplots(2, 2, figsize=(5, 5))
    axs[0, 0].imshow(img[:,:,0])
    axs[1, 0].imshow(img[:,:,1])
    axs[0, 1].imshow(img[:,:,2])
    #axs[1, 1].hist(img)
    # imgplot = plt.imshow(lum_img)
    return img
mimg = mat_version(file_name)
```

### 對彩色圖片使用matplotlib做統計，並顯示

```python
plt.hist(mimg.ravel(), bins=256, fc='k', ec='k')
```

## 使用Image工具包打開圖片

```python
im = Image.open(file_name)
```

### 轉成灰色

```python
grey_image = im.convert('L')
grey_image.show()
```

### 做成統計表

```python
gh = grey_image.histogram()
```
